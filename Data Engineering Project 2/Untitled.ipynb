```python
import pandas as pd
import sys
import requests
import io
import logging
import zipfile

# Set up logging for monitoring
logging.basicConfig(level=logging.INFO)

# Kaggle dataset URL and API key
dataset_url = "https://www.kaggle.com/api/v1/datasets/download/thedevastator/analyzing-credit-card-spending-habits-in-india"
headers = {'Authorization': 'Bearer <your-kaggle-api-key>'}

# Chunk size for processing large datasets (adjust based on memory constraints)
chunk_size = 10000  # Adjust this as necessary




def flatten_data(df):
    """Flatten the DataFrame if it contains lists in columns."""
    def safe_get_first_element(x):
        """Safely get the first element from a list or return the value if it's not a list."""
        if isinstance(x, list):
            return x[0]
        return x  # Return the value as is if it's not a list
    
    # Flatten columns by applying the safe_get_first_element function
    flattened_df = pd.DataFrame({
        "index": df["index"].apply(safe_get_first_element),
        "City": df["City"].apply(safe_get_first_element),
        "Date": df["Date"].apply(safe_get_first_element),
        "Card Type": df["Card Type"].apply(safe_get_first_element),
        "Exp Type": df["Exp Type"].apply(safe_get_first_element),
        "Gender": df["Gender"].apply(safe_get_first_element),
        "Amount": df["Amount"].apply(safe_get_first_element),
    })
    return flattened_df

try:
    # Start downloading dataset
    logging.info("Starting data download...")
    response = requests.get(dataset_url, headers=headers, stream=True)
    response.raise_for_status()  # Raise an error for bad responses

    # Unzip the downloaded file in memory and process the contents
    with zipfile.ZipFile(io.BytesIO(response.content)) as zip_ref:
        # List the files in the zip archive
        file_list = zip_ref.namelist()
        logging.info(f"Files in the zip archive: {file_list}")

        # Extract the CSV file
        with zip_ref.open(file_list[0]) as file:
            # Read data in chunks for memory-efficient processing
            logging.info("Processing the data in chunks...")
            chunks = []  # Initialize an empty list to collect chunks
            for chunk in pd.read_csv(file, chunksize=chunk_size):
                logging.info(f"Processed {len(chunk)} rows of data.")
                
                # Clean the chunk of data
                cleaned_chunk = chunk  # Apply cleaning function
                flattened_chunk = flatten_data(cleaned_chunk)  # Flatten the data
                chunks.append(flattened_chunk)  # Collect cleaned and flattened data

    # Combine all cleaned chunks into a single DataFrame
    final_df = pd.concat(chunks, axis=0)

    # Write the final DataFrame to Parquet and stream it via sys.stdout
    logging.info("Streaming the cleaned data to sys.stdout in Parquet format...")

    import pyarrow as pa
    import pyarrow.parquet as pq

    # Convert pandas DataFrame to PyArrow Table
    table = pa.Table.from_pandas(final_df)

    # Stream the Parquet file to stdout
#     pq.write_table(final_df, sys.stdout.buffer)  # Directly write to sys.stdout.buffer without BufferOutputStream
    final_df.to_parquet(sys.stdout.buffer)
    logging.info("Data successfully streamed to stdout in Parquet format.")
except requests.exceptions.RequestException as e:
    logging.error(f"Failed to download dataset: {e}")
except pd.errors.EmptyDataError:
    logging.error("No data found in the response.")
except zipfile.BadZipFile:
    logging.error("The downloaded file is not a valid zip file.")
except Exception as e:
    logging.error(f"An unexpected error occurred: {e}")
```

```python
import pandas as pd
import sys
import requests
import io
import logging
import zipfile
import pyarrow as pa
import pyarrow.parquet as pq

# Set up logging for monitoring
logging.basicConfig(level=logging.INFO)

# Kaggle dataset URL and API key
dataset_url = "https://www.kaggle.com/api/v1/datasets/download/thedevastator/analyzing-credit-card-spending-habits-in-india"
headers = {'Authorization': 'Bearer <your-kaggle-api-key>'}

# Chunk size for processing large datasets (adjust based on memory constraints)
chunk_size = 10000  # Adjust this as necessary

try:
    # Start downloading dataset
    logging.info("Starting data download...")
    response = requests.get(dataset_url, headers=headers, stream=True)
    response.raise_for_status()  # Raise an error for bad responses

    # Unzip the downloaded file in memory and process the contents
    with zipfile.ZipFile(io.BytesIO(response.content)) as zip_ref:
        # List the files in the zip archive
        file_list = zip_ref.namelist()
        logging.info(f"Files in the zip archive: {file_list}")

        # Extract the CSV file
        with zip_ref.open(file_list[0]) as file:
            # Read data in chunks for memory-efficient processing
            logging.info("Processing the data in chunks...")
            for chunk in pd.read_csv(file, chunksize=chunk_size):
                # Clean the data (modify this section as per your requirements)
                cleaned_chunk = chunk

                # Print cleaned data (for testing)
                logging.info(f"Processed {len(cleaned_chunk)} rows of data.")
                
                # Convert the pandas DataFrame to a PyArrow Table
                table = pa.Table.from_pandas(cleaned_chunk)

                # Create a buffer to write Parquet data
                buffer = io.BytesIO()

                # Write the PyArrow Table to the buffer in Parquet format
                pq.write_table(table, buffer, row_group_size=chunk_size)

                # Stream the buffer content to sys.stdout (ensure it's in binary)
                sys.stdout.buffer.write(buffer.getvalue())

    logging.info("Data processing and transmission to NiFi completed successfully.")

except requests.exceptions.RequestException as e:
    logging.error(f"Failed to download dataset: {e}")
except pd.errors.EmptyDataError:
    logging.error("No data found in the response.")
except zipfile.BadZipFile:
    logging.error("The downloaded file is not a valid zip file.")
except Exception as e:
    logging.error(f"An unexpected error occurred: {e}")
```
